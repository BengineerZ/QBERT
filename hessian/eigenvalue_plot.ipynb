```python
import pandas as pd 
import seaborn as sns; sns.set()
from matplotlib import pyplot as plt
import glob
%matplotlib inline
```

```python
df_full = pd.read_csv("results/CoLA-eigens.csv", delimiter = ',')
df_full["Percentage"] = 100
```

```python
sns.palplot(sns.color_palette("Paired", 12))
```

```python
plt.figure(figsize=(12, 12))
colors = sns.color_palette("Paired", 12)
fig = sns.lineplot(x="iters", y="max_eigenvalue", hue="block", palette=colors, data=df_full)
```

```python
df_samples = []
for filename in glob.glob('results/CoLA-10on100*-eigens.csv'):
    df_sample = pd.read_csv(filename, delimiter = ',')
    df_sample["Percentage"] = 10
    df_samples.append(df_sample)
```

```python
plt.figure(figsize=(12, 12))
colors = sns.color_palette("Paired", 12)
fig = sns.lineplot(x="iters", y="max_eigenvalue", hue="block", palette=colors, data=df_sample)
```

```python
merge_df = pd.concat(df_samples, axis=0)
```
